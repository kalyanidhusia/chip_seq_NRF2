```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   #Data visualisation libraries 
import seaborn as sns
from matplotlib.ticker import PercentFormatter
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
Positive_single_Nrf2 = pd.read_csv('Positive_singletonNRF2_info.csv')
Positive_single_Nrf2.head()
Positive_single_Nrf2.info()
Positive_single_Nrf2.describe()
Positive_single_Nrf2.columns
```

```python
sns.pairplot(Positive_single_Nrf2)
```

```python
X = Positive_single_Nrf2[['start_coord', 'end_coord',
                          'peak_assoc_with']]
y = Positive_single_Nrf2['peak_sum']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.5, random_state=1000)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```
